```python
import re 
import random
```

```python
class Rulebot:
    # potential negative word
    negative_responses = ("no","nope","nah","not a chance","sorry")
    #keyword for exiting the conversation
    exit_commands = ("quit","pause","exit","goodbye","bye","later")
    # random starter question
    random_questions = (
    "why are you here?",
    "are there many human like you",
    "what do you consume for sustenance",
    "Is there intelligent life on this planet",
    "Does Earth have a leader",
    "what planets have you visited",
    "what technology do you have on this planet",
    )
    
    def __init__(self):
        self.alinenbabble = {'describe_planet_intent': r'.*\s*your planet.*',
                            'answer_why_intent': r'why\sare.*',
                            'about_suraj':r'.*\s*suraj'
                             }
    
    def greet(self):
        self.name = input("what is your name?\n")
        will_help = input(
            f"Hi {self.name}, I am shisu. Will you help me to learn about your planet?\n")
        if will_help in self.negative_responses:
            print("OK, have a nice day!")
            return
        self.chat()
    
    def make_exit(self,reply):
        for command in self.exit_commands:
            if reply == command:
                print("Okey,have a nice day!")
                return True
        
    def chat(self):
        reply = input(random.choice(self.random_questions)).lower()
        while not self.make_exit(reply):
            reply = input(self.match_reply(reply))
        
    def match_reply(self, reply):
        for key, value in self.alinenbabble.items():
            intent = key
            regex_pattern = value
            found_match = re.match(regex_pattern, reply)
            if found_match and intent == 'describe_planet_intent':
                return self.describe_planet_intent()
            elif found_match and intent == 'answer_why_intent':
                return self.answer_why_intent()
            elif found_match and intent == 'about_suraj':
                return self.about_suraj()
        if not found_match:
            return self.no_match_intent() 
    
    def describe_planet_intent(self):
        responses = ("My planet name is a Bharat of diverse organisms and species.\n",
                     "I am from Jharkhand, the capital of Bokaro galaxie.\n")
        return random.choice(responses)
    
    def answer_why_intent(self):
        responses = ("I come in peace\n", "I am here to collect data of your planet and its inhabitants\n",
                     "I heard the tea is good\n")
        return random.choice(responses)

    def about_suraj(self):
        responses = ("I am suraj from jharkhand currenty living in Delhi.\n", "I am suraj a Data scientist and my skills is python,sql\n",
                     "I am suraj having also good command on power bi and ms excel\n")
        return random.choice(responses)

    def no_match_intent(self):
        responses = (
        "Please tell me more.\n", "Tell me more!\n", "why do you say that?\n", "I see. Can you elaborate?\n",
        "Interesting. Can you tell me more?\n", "I see.How do think?\n", "why?\n",
        "How do you think I feel when you say that?\n")
        return random.choice(responses)


Alienbot = Rulebot()
Alienbot.greet()

    


    
    
   
```
